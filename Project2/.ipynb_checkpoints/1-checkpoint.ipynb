```python
# def make_neuron_group(ex_num , inh_num , ex_p , inh,p):
start_scope()

u_rest = -70 * mV
threshold = -55 * mV
R = 3 * Mohm
tau = 8 * ms
I = 8 * namp
    
eqs = 'du/dt = -(u-u_rest)/tau + R*I/tau : volt'
G = NeuronGroup(1,eqs,threshold='u>threshold', reset='u=u_rest',method='linear')
G.u = u_rest
M = StateMonitor(G, 'u', record=0)

run(30*ms)
```

```python
plot(M.t/ms, M.u[0]/mV)
xlabel('Time (ms)')
ylabel('v (mv)');
```
